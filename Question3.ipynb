```python
import pandas as pd

# Load the dataset
df = pd.read_csv('tourist_attractions.csv')

import folium

# Create a map centered around Malaysia
map_malaysia = folium.Map(location=[4.2105, 101.9758], zoom_start=6)

# Function to add markers to the map
def add_markers(map_obj, df):
    for _, row in df.iterrows():
        folium.Marker(
            location=[row['latitude'], row['longitude']],
            popup=f"<strong>{row['name']}</strong><br>{row['description']}",
            icon=folium.Icon(color='blue')
        ).add_to(map_obj)

# Add markers to the map
add_markers(map_malaysia, df)

# Display the map
map_malaysia.save('malaysia_tourist_map.html')
map_malaysia
```

C.

```python
from folium.plugins import FeatureGroupSubGroup

# Define a color dictionary based on attraction type
color_dict = {
    'historical site': 'red',
    'natural wonder': 'green',
    'amusement park': 'blue'
}

# Create feature groups for different types of attractions
historical_sites = folium.FeatureGroup(name='Historical Sites')
natural_wonders = folium.FeatureGroup(name='Natural Wonders')
amusement_parks = folium.FeatureGroup(name='Amusement Parks')

# Function to add markers with different colors
def add_custom_markers(map_obj, df):
    for _, row in df.iterrows():
        marker = folium.Marker(
            location=[row['latitude'], row['longitude']],
            popup=f"<strong>{row['name']}</strong><br>{row['description']}",
            icon=folium.Icon(color=color_dict.get(row['type'], 'blue'))
        )
        if row['type'] == 'historical site':
            marker.add_to(historical_sites)
        elif row['type'] == 'natural wonder':
            marker.add_to(natural_wonders)
        elif row['type'] == 'amusement park':
            marker.add_to(amusement_parks)

# Add custom markers to the map
add_custom_markers(map_malaysia, df)

# Add feature groups to the map
historical_sites.add_to(map_malaysia)
natural_wonders.add_to(map_malaysia)
amusement_parks.add_to(map_malaysia)

# Add layer control
folium.LayerControl().add_to(map_malaysia)

# Save the customised map
map_malaysia.save('custom_malaysia_tourist_map.html')
map_malaysia
```
